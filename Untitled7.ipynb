```python
%matplotlib widget
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set up plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.set_ylim([-4, 4])
ax.grid(True)
 
# generate x values
x = np.linspace(0, 2 * np.pi, 100)

 
def my_sine(x, w, amp, phi):
    """
    Return a sine for x with angular frequeny w and amplitude amp.
    """
    return amp*np.sin(w * (x-phi))
 
 
@widgets.interact(w=(0, 10, 1), amp=(0, 4, .1), phi=(0, 2*np.pi+0.01, 0.01))
def update(w = 1.0, amp=1, phi=0):
    """Remove old lines from plot and plot new one"""
    [l.remove() for l in ax.lines]
    ax.plot(x, my_sine(x, w, amp, phi), color='C0')
```

```python
# Height in meters
height = 50.0

# Rope length in meters
rope_length = 20.0

# Rope stiffness
rope_k = 100.0

# Person mass in kilogramm
# mass = 80

# Gravitational constant
gravity = 9.81

# Internal friction + air friction
friction_coefficient = 20.0
```

```python
# Define Newton's law of motion
def F(x, v, mass_in, gravity_in, rope_k_in, height_in, rope_length_in, friction_coefficient_in):
  return - mass_in * gravity_in + rope_k_in * np.max([0.0, height_in - x - rope_length_in]) - friction_coefficient_in * v

# Define PDE
f0 = np.array([height, 0.0])

def f(fx, t, mass_in, gravity_in, rope_k_in, height_in, rope_length_in, friction_coefficient_in):
  return np.array([fx[1],1.0 / mass_in * F(fx[0],fx[1],mass_in, gravity_in, rope_k_in, height_in, rope_length_in, friction_coefficient_in)])
```

```python
tmin = 0.0
tmax = 25.0
nt = 5000

tcoord = np.linspace(tmin,tmax,nt+1)
print('Domain discretized with dt = '+str(tcoord[1]-tcoord[0]))
```

```python
def grapher(mass_in, gravity_in, rope_k_in, height_in, rope_length_in, friction_coefficient_in):
  sol = odeint(f, f0, tcoord, args=(mass_in,gravity_in,rope_k_in,height_in,rope_length_in,friction_coefficient_in))
  Ekin = 0.5 * mass_in * np.transpose(sol)[1]**2.0
  Epot = mass_in * gravity_in * np.transpose(sol)[0]
  Espring = 0.5 * rope_k * np.maximum((height_in - np.transpose(sol)[0] - rope_length_in)*0.0, height_in - np.transpose(sol)[0] - rope_length_in)**2.0
  
  return np.array([np.transpose(sol)[0], Ekin, Epot, Espring])

```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(7,7), dpi=96)

ax1.set_aspect(0.5*0.5)
ax1.set_xlim([0,25])
ax1.set_ylim([0,50])
ax1.grid()

ax2.set_xlabel('Time (seconds)')
ax2.set_ylabel('Height (meters)')
ax2.set_xlabel('Time (seconds)')
ax2.set_ylabel('Energy (joule)')

ax2.set_xlim([0,25])
ax2.set_ylim([0,50000])
ax2.set_aspect(0.5*25/50000.0)
ax2.grid()

@widgets.interact(mass=(10,100,1))
def update(mass = 80):
    [l.remove() for l in ax1.lines]
    ax1.plot(tcoord,grapher(mass,gravity,rope_k,height,rope_length,friction_coefficient)[0])
    ax2.plot(tcoord,grapher(mass,gravity,rope_k,height,rope_length,friction_coefficient)[1], label=r'$E_{\rm kin}$')
    ax2.plot(tcoord,grapher(mass,gravity,rope_k,height,rope_length,friction_coefficient)[2], label=r'$E_{\rm pot}$')
    ax2.plot(tcoord,grapher(mass,gravity,rope_k,height,rope_length,friction_coefficient)[3], label=r'$E_{\rm spring}$')
    ax2.plot(tcoord,grapher(mass,gravity,rope_k,height,rope_length,friction_coefficient)[0]+
                grapher(mass,gravity,rope_k,height,rope_length,friction_coefficient)[1]+
                grapher(mass,gravity,rope_k,height,rope_length,friction_coefficient)[2], label=r'$E_{\rm tot}$')
    
    ax2.legend()
```
